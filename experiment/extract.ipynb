# Experiment Results Summary

```python
# 현재 디렉토리 내에 있는 txt 파일을 모두 읽어서 data 변수에 리스트로 저장
import os
data_cora = []
data_citeseer = []
data_pubmed = []
for filename in os.listdir('.'):
    if filename.endswith('.txt'):
        with open(filename) as file:
            # 첫줄을 읽고 그 내용에 따라 다른 변수에 저장
            first_line = file.readline()
            if first_line.startswith('Cora'):
                data_cora.append(file.read())
            elif first_line.startswith('Citeseer'):
                data_citeseer.append(file.read())
            elif first_line.startswith('Pubmed'):
                data_pubmed.append(file.read())
            else:
                print('Unknown dataset:', first_line)
```

```python
results_cora = {
    'GAT': [[],[],[],[]],
    'GATRWR': [[],[],[],[]],
    'GATIRC': [[],[],[],[]],
    'GATIRCRWR': [[],[],[],[]]
}

results_citeseer = {
    'GAT': [[],[],[],[]],
    'GATRWR': [[],[],[],[]],
    'GATIRC': [[],[],[],[]],
    'GATIRCRWR': [[],[],[],[]]
}

results_pubmed = {
    'GAT': [[],[],[],[]],
    'GATRWR': [[],[],[],[]],
    'GATIRC': [[],[],[],[]],
    'GATIRCRWR': [[],[],[],[]]
}

datas = [data_cora, data_citeseer, data_pubmed]
results = [results_cora, results_citeseer, results_pubmed]

for i in range(len(datas)):
    print(i)
    for d in datas[i]:
        print(datas[i])
        lines = d.split('\n')
        model = ""
        layer_cnt = 0
        for l in lines:
            # 만약 l 이 layers라는 문자열을 포함하고 있다면
            if 'layers' in l:
                layer_cnt = int(l[0])
                if l.endswith('GAT'):
                    model = "GAT"
                elif l.endswith('GAT with Random walk with restart'):
                    model = "GATRWR"
                elif l.endswith('GAT with Initial residual connection'):
                    model = "GATIRC"
                elif l.endswith('GAT with Random walk with restart and Initial residual connection'):
                    model = "GATIRCRWR"
            elif 'Test accuracy' in l:
                results[i][model][layer_cnt-2].append(float(l.split("= ")[-1]))
```
